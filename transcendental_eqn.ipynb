```python
import numpy as np
import math

x = []
er = []
cnt = 0
for i in np.arange(0.1, 5, 0.001):
    x.append(i)
    er.append(math.erf(x[-1]))
    cnt = cnt+1
    
x = np.array(x)
er = np.array(er)
y =[]
y = np.exp(-x*x)/(er*math.sqrt(math.pi)*x)-2

for i in range(1,cnt):
#     print(i)
    if(y[i]*y[i-1]<0):
        print(x[i])
```
